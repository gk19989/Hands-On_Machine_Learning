## 1) Does money bring happiness?

In the first section, we shall try to answer a common philosophical question **"Does money bring happiness?"**

To answer this question we are considering two datasets:
 1. Better Life Index data from Organisation for Economic Co-Operation and Development(OECD)
 2. GDP per capita data from International Monetory Fund(IMF)

```python
# Making space for our datasets

import os
datapath = os.path.join("datasets", "lifesat","")
```

```python
# To plot pretty firgures directly inside jupyter

%matplotlib inline
import matplotlib as mpl
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

We shall download the data from "Hands-On Machine Learning with Scikit-Learn, Keras, and TensorFlow, 2nd Edition's" github repo 

```python
# Download the data

import urllib
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
os.makedirs(datapath, exist_ok = True)
for filename in ("oecd_bli_2015.csv", "gdp_per_capita.csv"):
    print("Downloading", filename)
    url = DOWNLOAD_ROOT + "datasets/lifesat/" + filename
    urllib.request.urlretrieve(url, datapath + filename)
```

### Data preprocessing or preparation

```python
import pandas as pd
```

```python
oecd_bli = pd.read_csv(datapath + "oecd_bli_2015.csv", thousands=',')
```

```python
oecd_bli
```

```python
gdp_per_capita = pd.read_csv(datapath + "gdp_per_capita.csv",thousands=',',delimiter='\t',
                             encoding='latin1', na_values="n/a")
```

```python
gdp_per_capita
```

```python
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
```

```python
oecd_bli
```

Taking only what is necessary

```python
oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
```

```python
oecd_bli
```

```python
gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
```

```python
gdp_per_capita
```

```python
gdp_per_capita.set_index("Country", inplace=True)
```

```python
gdp_per_capita
```

Merging the two tables

```python
full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita, left_index=True, right_index=True)
```

```python
full_country_stats
```

```python
full_country_stats.sort_values(by="GDP per capita", inplace=True)
```

### Time to apply ML

We shall apply LinearRegression Model between **GDP** and **Life satisfaction** and try to see if there is any relation between them

```python
# imports

import matplotlib.pyplot as plt
import numpy as np
import sklearn.linear_model
```

```python
X = np.c_[full_country_stats["GDP per capita"]]
Y = np.c_[full_country_stats["Life satisfaction"]]
```

```python
full_country_stats.plot(kind='scatter', x="GDP per capita", y="Life satisfaction")
plt.show()
```

```python
model = sklearn.linear_model.LinearRegression()
```

```python
model.fit(X, Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
X_new = [[22587]]
print(model.predict(X_new))
```

```python
plt.scatter(X, Y, color='g')
plt.plot(X, model.predict(X), color='k')
plt.show()
```

**Now you can clearly see there is linear relation between GDP per capita and Life Satisfaction. This means now with the assurance of data we can say that ***MONEY DOES BRING HAPPINESS*
